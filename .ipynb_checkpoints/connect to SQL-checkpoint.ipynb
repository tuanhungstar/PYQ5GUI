# SQL Function

```python
def find_product_group(keyword):
    sql = "SELECT * FROM ctxf.dbo.hscodedata where [PRODUCT GROUP] like '%" + str(keyword)+ "%'"
    df_out = pd.read_sql(sql, conn)
    return df_out

def get_col_name(qsl_table):
    sql = "SELECT * FROM ctxf.dbo." + qsl_table + " where [Term code] like '%" + 'NOTWORD'+ "%'"
    df = pd.read_sql(sql, conn)
    col_name ='' 
    Value = ''
    for col in df.columns:
        if len(col)>2:
            col_name = col_name + '[' + str(col) +'],'
            Value = Value + "?,"
    Value = '(' + Value[0:len(Value)-1]+ ")"
    col_name = '(' + col_name[0:len(col_name)-1]+ ")"
    return col_name,Value ,df.columns.values
def write_to_sql(df,col_name,Value,table_name):
    sql = "INSERT INTO ctxf.dbo." + table_name + " " +  col_name  + ' Values ' + Value 
    val = tuple(df.values)
    array_of_tuples = map(tuple, val)
    tuple_of_tuples = tuple(array_of_tuples)
    val = tuple_of_tuples
    cursor.executemany(sql,val)
    conn.commit()
    return

def write_to_sql_list_file(filename,status,comment):
    now = datetime.now()
    sql = "INSERT INTO ctxf.dbo.Listfileadd (File_Name,Add_To_SQL_Status,Add_Date,Comment) Values(?,?,?,?)"
    val = (filename,status,str(now),comment)
    cursor.execute(sql,val)
    conn.commit()
    return

def remove_duplicate_pn():
    sql_remove_duplicate = '''WITH cte AS (
                                SELECT 
                                    [Product Number], 

                                    ROW_NUMBER() OVER (
                                        PARTITION BY 
                                            [Product Number] 
                                        ORDER BY 
                                            [Product Number] 
                                    ) row_num
                                 FROM [ctxf].[dbo].[PEP_file_all]
                            )
                            DELETE FROM cte
                            WHERE row_num > 1'''
    cursor.execute(sql_remove_duplicate)
    conn.commit()
    return
def remove_duplicate_rule():
    sql_remove_duplicate = '''WITH cte AS (
                                            SELECT 
                                               [Number Range]
                                              ,[Term code]
                                              ,[Key Word]
                                              ,[Reference Number]
                                              ,[Number of Part Used], 

                                                ROW_NUMBER() OVER (
                                                    PARTITION BY 
                                                        [Number Range],
                                                        [Term code],
                                                        [Key Word],
                                                        [Reference Number]
                                                    ORDER BY 
                                                        [Number Range],
                                                        [Term code],
                                                        [Key Word],
                                                        [Reference Number]
                                                ) row_num
                                             FROM [ctxf].[dbo].[RULEBASE]
                                        )
                                        DELETE FROM cte
                                        WHERE row_num > 1;'''
    cursor.execute(sql_remove_duplicate)
    conn.commit()
    return
def Delete_all_data_in_table(table_name):
    sql = 'TRUNCATE TABLE [ctxf].[dbo].[' + table_name + '];'  
    cursor.execute(sql)
    conn.commit()
    return
def Load_all_File_SQL(Folder_name,SQL_table):
    col_name,Value,use_col = get_col_name(SQL_table)
    File_save_ok = []
    File_save_Not_ok = []
    for file in get_list_file(Folder_name):
        try:
            df=pd.read_excel(Folder_name + file, 'Sheet1',usecols=use_col)
            df=df[use_col] # re-order columns to align with SQL data base
            df=df.fillna(int(0))
            if df['Product Number'].count()>0:
                write_to_sql(df,col_name,Value,SQL_table)
                write_to_sql_list_file(file,'OK','good')
                File_save_ok.append(file)
            else:
                write_to_sql_list_file(file,'NG','Error')
                File_save_Not_ok.append(file)
        except:
            File_save_Not_ok.append(file)
            
    return 
```

# Classify by Term code and Number Range only

```python
def classify_by_term(row):
    classify_by_term = Rule_list[Rule_list['Term code']==row['Term code']]
    classify_by_range = Rule_list[Rule_list['Number Range']==row['Product Number'][0:4]]
    classify_by_term_range = Rule_list[(Rule_list['Number Range'].str.startswith(row['Product Number'][0:4])) & (Rule_list['Term code'].str.startswith(row['Term code']))]
    if classify_by_term_range['Reference Number'].count()==1:  
        part_ref = classify_by_term_range.iloc[0]['Reference Number']#.values ['Reference Number']
        Total_used = classify_by_term_range.iloc[0]['Number of Part Used']    
        rule_name = 'C'
    elif classify_by_term['Reference Number'].count()==1:
        part_ref = classify_by_term.iloc[0]['Reference Number']#.values ['Reference Number']
        Total_used = classify_by_term.iloc[0]['Number of Part Used']
        rule_name = 'B'
    elif classify_by_range['Reference Number'].count()==1:
        part_ref = classify_by_range.iloc[0]['Reference Number']#.values ['Reference Number']
        Total_used = classify_by_range.iloc[0]['Number of Part Used']
        rule_name = 'A'
    else:
        part_ref=''
        Total_used=0
        rule_name=''
    return part_ref,Total_used,rule_name
```

# Classify 1 part by term code number range and description

```python
def get_ref_from_possible(df_posible):

    #if there are more than 1 posbible
    if len(df_posible.index) >1: 
        #2nd posible have less them 30% total part used
        if (df_posible.iloc[1:2,0:1].values[0][0]/df_posible.iloc[0:1,0:1].values[0][0])<=0.3: 
            ref= df_posible.index[0]
            number_used = df_posible.iloc[0:1,0:1].values[0][0]
            rule_name ="D"
        else:
            ref= 'Conflicted'
            number_used = 0
            rule_name =""
        #if only 1 possible
    elif len(df_posible.index==1):
            ref= df_posible.index[0]
            number_used = df_posible.iloc[0:1,0:1].values[0][0]
            rule_name ="D"
        #if no posbile
    else:
        ref= 'No rule found'
        number_used = 0 
        rule_name =""
    return ref,number_used,rule_name      

def classify_1_part_by_nr_tc_kw(partnumber,part_des,termcode):
    global df_with_des
    #df1 = df[df['Product Number'] == partnumber]
    #termcode = df1['Term code'].values[0]
    #part_des = df1['Matl Type'].values[0]
    nRange = partnumber[0:4]
    df_matched=df_with_des[(df_with_des['Term code'] == termcode) & (df_with_des['Product Number'].str[0:4] == nRange)].copy()
    if df_matched.count()[0]>0:
        # add col fuzzy matched poit
        df_matched['match point'] = df_matched.apply(lambda row: add_match_point(part_des,row), axis=1) 
        df_matched.sort_values(by='match point', ascending=False,inplace=True) # sort by match point column
        df_filtered_match_point = df_matched[df_matched['match point']>=70] # threshold value of match point will pickup

        # get list of ref parts are posible for this case
        df_posible = df_filtered_match_point.groupby('Reference Number').count().sort_values(by='Logical System Group', ascending=False)
        result = get_ref_from_possible(df_posible)
    else:
        result='',0,''
    return result
```
